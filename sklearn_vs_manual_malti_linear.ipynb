```python
import numpy as np
import pandas as pd 
```

```python
df=pd.read_csv("D:\hart\heart.csv")
df.head(4)
```

```python
df.columns
```

```python
x=df[['Age','RestingBP','MaxHR']].values
x=np.c_[np.ones((x.shape[0],1)),x]
```

```python
y=df['Cholesterol']
```

```python
x_transpose=x.T
```

```python
w=np.linalg.inv((x_transpose.dot(x))).dot(x_transpose).dot(y)
```

```python
print("intercept =",w[0])
print("slope m1 =",w[1])
print("slope m2 =",w[2])
print("slope m3 =",w[3])
```

```python
new_val=np.array([1,40,120,172])
```

```python
y_predict=new_val.dot(w)
```

```python
print(y_predict)
```
